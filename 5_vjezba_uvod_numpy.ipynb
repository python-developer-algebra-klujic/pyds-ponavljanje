Kreirajte funkciju, naziva *funkcija*, koja prima dva argumenta i vraca njihov umnozak s brojem 5

```python
def funkcija(x, y):
    return x * y * 5


np.fromfunction(funkcija, (5, 8), dtype=np.int64)
```

Prepravite gornji primjer tako da ispisuje tablicu mnozenja do 15

```python
def tablica_mnozenja(x, y):
    return x * y


nd_tablica_mnozenja = np.fromfunction(tablica_mnozenja, (16, 16), dtype=np.int64)
nd_tablica_mnozenja = nd_tablica_mnozenja[ 1: , 1:]
nd_tablica_mnozenja
```

```python
nd_tablica_mnozenja_do_10 = nd_tablica_mnozenja[ :10 , :10 ]
nd_tablica_mnozenja_do_10
```

```python
nd_tablica_mnozenja_do_10[4:6, 4:6] = 99999
nd_tablica_mnozenja_do_10
```
