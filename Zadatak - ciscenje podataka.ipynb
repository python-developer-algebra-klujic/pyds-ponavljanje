3. Korak - Primjena statistike za incijalnu analizu podataka ako se radi o brojevima, a ako se radi o tekstu, onda primjena string metoda.

```python
import statistics


def statictic_alnalyze(data: list):
    minimalna_visina = min(data)
    maksimalna_visina = max(data)
    # prosjecna_visina = round((sum(visine_hrasta) / len(visine_hrasta)), 2)
    prosjecna_visina = round((statistics.mean(data)), 2)
    median_visina = round((statistics.median(data)), 2)
    
    print(f'Minimalna vrijednost: {minimalna_visina}')
    print(f'Maksimalna vrijednost: {maksimalna_visina}')
    print(f'Prosjecna vrijednost: {prosjecna_visina}')
    print(f'Median: {median_visina}')


statictic_alnalyze(visine_hrasta)
```

```python
# List slice notation - vraca kolkeciju vrijednosti definiranih indeksima
# list_name[ start_index : stop_index : step ]
# kao kod range() - start je ukljucen, stop NIJE ukljucen

# visine_hrasta[0 : 10 : 1]
visine_hrasta[ : 10] # ako ne napisemo start_indeks onda se podrazumijeva da je to 0
```

```python
visine_hrasta[-1]
```

```python
visine_hrasta[ -3 : -1]
```

```python
visine_hrasta[ -3 : 0]
```

```python
visine_hrasta[ -3 : ] # na ovaj nacin cemo ukljuciti i zadnji element
```

```python
visine_hrasta[ : ] # vratiti ce sve elemente od prvog do zadnje ukljucivo
```

```python
visine_hrasta[ -3 : 0] # od od indeksa -3 prema zanjem elementu, odnosno prema desno, odnosno u + smjeru, jer je korak +1
```

```python
visine_hrasta[ -3 : 0 : -1] # sada ide od predzadnjeg elementa prema prvom (koji nece biti ukljucen), ali u - smjeru, odnosno od desna prema lijevo
```

```python
visine_hrasta[ -3 :  : -1] # sada ce ukljuciti i prvi element
```

```python
# ZADATAK
# Pomocu slice notacije vratite listu u poredanu od zdanjeg do prvog elementa

# RJESENJE
visine_hrasta[ -1 : : -1 ]
```
