# Korištenje Pandasa

U nastavku prikazat ćemo neke naprednije mogućnosti Pandasa kroz primjere. Učitajmo podatke za Airbnb.

```python
import pandas as pd

airbnb_data = pd.read_csv('airbnb_new_york_city_listings.csv')

airbnb_data
```

Prikaz u Jupyter Notebook-u koristi `display` funkciju, koju ne moramo posebno pozivati. Ovdje pokazujemo da postoji i da se poziva implicitno.

```python
display(airbnb_data) # Isto kao da je zadnja linija `airbnb_data`
```

Za početak analizirajmo kojeg su tipa podataka stupci.

```python
airbnb_data.dtypes
```

Stupac `last_review` je tipa `object`, a iz podataka i samog naziva stupca vidimo da bi to trebao biti datum. Pokušajmo pretvoriti cijeli taj stupac u `DateTime` tip podatka.

```python
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'])
airbnb_data.dtypes
# Alternativno: airbnb_data['last_review'] = airbnb_data['last_review'].astype('datetime64')
```

Kako je `datatime64` tip podatka iz kojeg možemo dobiti samo godinu, to ćemo iskoristiti tako što ćemo stvoriti novi stupac naziva `year`.

```python
airbnb_data['year'] = airbnb_data['last_review'].dt.year
airbnb_data['year']
```

Provjerimo ako neki podaci nedostaju.

```python
airbnb_data.isnull().sum()
```

Imamo pet kolona koje imaju prazne ćelije. Kolone "name" i "host_name" su tipa object, odnosno tekstualnog tipa, dok su "reviews_per_month" i "last_review_year" decimalni brojevi, a kolona "last_review" je tipa *DateTime*.

Unesimo podatke u prazne ćelije tako da nam kasnije ne utječu na proračune. To će najbolje biti ako za kolonu:
- "name": "No Name Listing"
- "host_name": "J. Doe"
- "reviews_per_month": 0.00
- "last_review_year": 0
- "last_review": '01-01-1970', što bi odgovaralo početku UNIX vremena

Za ovu aktivnost imamo na raspolaganju tri metode:
- *fillna()*
- *replace()*
- *interpolate()*

Mi ćemo koristiti metodu *fillna()*. Kad koristimo `inplace` argumet `True`, ne stvara se kopija, već se podaci promjene na originalnom objektu.

Više detalja o navedenim funkcijama možete pronaći na stranici: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html.

```python
airbnb_data['name'].fillna('No Name Listing', inplace=True)
airbnb_data['host_name'].fillna('J. Doe', inplace=True)
airbnb_data['reviews_per_month'].fillna(0.00, inplace=True)
airbnb_data['year'].fillna(0.00, inplace=True)
airbnb_data['last_review'].fillna('01-01-1970', inplace=True)
```

Stupac `year` je `float` tip podatka. Godine su cijeli brojevi, pa ćemo napraviti i tu transformaciju.

```python
airbnb_data.dtypes
```

`last_review` je opet postao `object` tip. Promjenimo i taj stupac zajedno sa `year`.

```python
airbnb_data['year'] = airbnb_data['year'].astype('int64')
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'])
airbnb_data.dtypes
```

Prikazat ćemo trenutno stanje našeg `DateFrame`-a.

```python
airbnb_data
```

Odlično, imamo jednu kolonu više od inicijalnog skupa podataka, nemamo praznih ćelija.

Većinu tipova podataka smo očistili, međutim, često se dogodi da nam tekstualni podaci sadrže znakove koje računala ne prikazuju (primjer je znak za novi red '*\n*'). Ti znakovi su tu, no kako nam ne trebaju u podacima, bilo bi ih dobro obrisati. Pandas podržava string metodu *str.strip()* nad *DataFrame* objektom. Ova metoda briše sve posebne znakove, prazna mjesta (razmaknicu ili *space*) te slične znakove na početku i na kraju tekstualnog podatka.

Očistimo i tekstualne podatke pomoću *str.strip()* metode, u svakoj koloni koja ima tekstualni tip podatka.

*INFORMACIJA: možemo koristiti i str.lower() i str.upper() metode ako nam trebaju. U našem slučaju one nisu korisne. Više informacija možete pronaći na stranici: https://pandas.pydata.org/docs/reference/api/pandas.Series.str.strip.html.*

```python
airbnb_data['name'] = airbnb_data['name'].str.strip()
airbnb_data['host_name'] = airbnb_data['host_name'].str.strip()
airbnb_data['neighbourhood_group'] = airbnb_data['neighbourhood_group'].str.strip()
airbnb_data['neighbourhood'] = airbnb_data['neighbourhood'].str.strip()
airbnb_data['room_type'] = airbnb_data['room_type'].str.strip()

airbnb_data
```

Prikažimo imena svih stupaca.

```python
airbnb_data.columns
```

Vidimo da imamo kolone naziva `price` i `minimum_nights`. Jedna kolona prikazuje cijenu najma po noćenju, a druga prikazuje koliko se minimalno noćenja mora uzeti. Ako spojimo ove dvije kolone, možemo dobiti novu koja će nam prikazati podatak o minimalnom prihodu. Dakle, ako pomnožimo cijenu s minimalnim brojem noćenja, onda je to minimalni prihod koji iznajmljivač može dobiti za svoju nekretninu.

Ako nova kolona prikazuje podatak o minimalnom prihodu, onda ćemo je nazvati `minimal_revenue`.

```python
airbnb_data['minimal_revenue'] = airbnb_data['price'] * airbnb_data['minimum_nights']
airbnb_data
```

Uzmimo samo tri stupca: `room_type`, `price`, `minimal_revenue`. Grupirajmo podatke prema tipu sobe: `room_type`. A nakon toga prema tipu sobe i godini.

```python
airbnb_data[['room_type', 'price', 'minimal_revenue']].groupby('room_type').mean()
```

```python
airbnb_data[['room_type', 'year', 'price']].groupby(['room_type', 'year']).median()
```

## Korelacija

Korelacija pokazuje ovisnost izmđu skupova podataka. Strogo gledano, korelacija pokazuje linearnu povezanost. Npr. ako proučavamo dva podatka, ocjenu učenika i vrijme učenja, očekujemo da će ti podaci biti korelirani. Razlog tome je što ako učenik potroši više vremena na učenje može očekivati veću ocjenu. Treba biti oprezan oko korelacije, jer neki podaci su korelirani, ali ne linearno!

Ako je rezultat korelacije 1, to znači da je veza između podataka savršeno linearna. Ako je korelacija jednaka 0, to znači da podaci uopće nisu korelirani (nema linearne veze). Ako je korelacija -1 podaci su savršeno obrnuto linarno povezani. Dobar primjer korelacije -1 bi bila veza između ocjene i vremena provedenog igraćuji video igre.

Izračunajmo korelaciju između stupaca učitanih i očišćenih podataka.



```python
airbnb_data.corr()
```

Zanima nas korelacija samo za stupac `price` i sve ostale.

```python
airbnb_data.corr()['price']
```

```python
airbnb_data.corr()['price'].sort_values()
```

## Pivot tablice

Pivot tabele ste vjerojatno upoznali u Microsoft Excelu. Njihova primjena je sumirati, sortirati, reorganizirati, grupirati, prebrojati sve ili prosjek podataka u tabeli.

Možda ovo pomogne boljem razumijevanju pivot tabele. Pivot tabela može:
- grupirati pojedinačne zapise, redove u kategorije
- prebrojati elemente u svakoj kategoriji
- sumirati vrijednosti elemenata
- izračunati srednju vrijednost, minimum ili maksimum.

Pandas ima ugrađenu metodu *pivot_table()*. Upoznajmo se s njezinim mogućnostima. Više detalja na: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html?highlight=pivot_table#pandas.pivot_table.

Kad radimo pivotiranje, moramo navesti četiri parametra:

- `values`: podaci koje želimo agregirati
- `index`: redci koje želimo u pivot tablici
- `columns`: stupci koje želimo u pivot tablici
- `aggfunc`: funkcija kojom želimo agregirati - statistika


```python
pd.pivot_table(airbnb_data, values=['price'], index=['room_type'], columns=['year'], aggfunc='median')
```
