#### Grananje

Kako se možemo granati u programu?

Ako varijabla *lista* sadrži više od 2 elementa ispišite "Lista je velika", ako lista sadrži između 2 i 5 elemenata "Lista je srednja", inače "Lista je mala"

```python
if len(lista) > 2:
    print("Lista je velika")
elif len(lista) >=2 and len(lista) <= 5:
    print("Lista je srednja")
else:
    print("Lista je velika")
```

#### Moduli

```python
import random
from random import randint
from random import randint as rnd

lista = []

for _ in range(10):
    lista.append(random.randint(0, 10))
    lista.append(randint(0, 10))
    lista.append(rnd(0, 10))

print(lista)
lista.sort()
print(lista)
```

#### Funkcije

Dio programskog koda koji se izvršava na poziv.

```python
def srednja_vrijednost(lista_brojeva):
    N = len(lista_brojeva)
    if N == 0:
        raise Exception("Lista je prazna")
    
    suma = 0
    for i in lista_brojeva:
        suma += i
    
    return suma/N

r1 = srednja_vrijednost([100, 1, 2, 30, 40]) # Poziv funkcije
r2 = srednja_vrijednost([1, 2, 30, 40]) # Poziv funkcije

print(r1, r2)

# Sto ako je lista prazna? Ima li jos slucajeva koje moramo pokriti?
```
