#### Datoteke

 Što je datoteka?

 Resurs koji sadrži podatke. Nalazi se u spremniku računala.
 Datoteke se najčešće koriste kao trajni spremnik podataka.

 Za razliku od varijabli u programu, koje nestaju kada se program prestaje s radom, podaci u datoteci ostaju očuvani. 

```python
datoteka = open('test.txt', 'w') # otvaranje datoteke, varijabla `datoteka` sluzi za upravljanje
datoteka.write('Ovaj tekst smo ubacili u datoteku')
datoteka.close() # datoteku moramo i zatvoriti jer koristimo resurse operacijskog sustava!
```

Iz datoteka naravno možemo i čitati podatke.

```python
datoteka = open('test.txt', 'r') # otvaranje datoteke za citanje, varijabla `datoteka` sluzi za upravljanje
sadrzaj = datoteka.read()
datoteka.close() # datoteku moramo i zatvoriti jer koristimo resurse operacijskog sustava!

sadrzaj
```

##### Vrste datoteka

Datoteke se dijele na **binarne** i **tekstualne**.
Jednu tekstualnu datoteku smo već kreirali i nakon toga pročitali.
Binarne datoteke u sebi sadrže proizvoljni binarni zapis podataka. U Pythonu se prilikom otvaranja binarne datoteke koristi zastavica `b`.

```python
datoteka = open('binarna_datoteka.dat', 'wb')
datoteka.write(bytearray([0xFF, 0x00, 0x7F, 0x80])) # zapisujemo bytove direktno
datoteka.close()
```

U podatkovnoj znanosti, bitni su formati datoteka. Za neke formate datoteka smo već čuli: docx, pdf, mp3, jpg, html, ...
Format datoteke propisuje kako su podaci strukturirani unutar datoteke. U podatkovnoj znanosti formati datoteka su bitni da bi smo mogli iz datoteka pročitati same podatke i s njima nešto raditi.

Jedan od poznatijih format je csv (comma-separated value). To je format podatka gdje su podaci organizirani u tablicu, a međusobno su odvojeni zarezom (a mogu biti i nekim drugim znakom ,; TAB).

```python
polaznici = """Ime,Prezime,Ocjena
Pero,Peric,5
Ivan,Ivic,3
Ana,Anic,4
Leo,Leic,1
"""
datoteka = open('polaznici.csv', 'w')
datoteka.write(polaznici)
datoteka.close()
```

Python sadrži ugrađene module, pa ćemo iskoristiti jedan koji nam omogućuje lakše upravljanje csv datotekama.
Za potrebe podatkovne znanosti postoje još napredniji modulu koje ćemo raditi kasnije.

```python
import csv # Uključujemo modul u naš program

with open('polaznici.csv') as csvfile: # Čemu služi with ključna riječ?
    polaznici = csv.reader(csvfile, delimiter=',')
    for row in polaznici:
        print(row) # Svaki redak datoteke postao je lista!
```

#### Klase i objekti

Python je objektno orijentirani programski jezik, ali koristi i druge paradigme.

Kad programiramo u Pythonu, zapravo radimo s objektima. Sve što koristimo u Pythonu je objekt.
A što je objekt?
Objekt je dio memorije nad kojim možemo raditi određene operacije. Npr. lista [1, 2, 3] je objekt, jer sadrži određene podatke u memoriji, a nad listom možemo raditi operacije kao što su sortiranje, ubacivanje, brisanje elemmenta...

A što je klasa?

Klasa je "predložak" po kojem kreiramo objekte. Npr. liste [1, 2, 3] i ['a', 'b', 10] su dva objekta, ali to su objekti iste klase, a to je lista.

Kreirajmo klasu `Osoba`:

```python
class Osoba:

    def __init__(self, ime, prezime, godina):
        self.ime = ime
        self.prezime = prezime
        self.godina = godina

    def stariji_od(self, druga_osoba):
        return self.godina > druga_osoba.godina
    
osoba_1 = Osoba("Ivan", "Ivic", 27)
osoba_2 = Osoba("Ante", "Antic", 30)

print(osoba_1.stariji_od(osoba_2))
```

U predhodnom kodu pokazali smo kako kreirati klasu `Osoba`. Klasa osoba ima konstruktor `__init__`, koji zahtjeva pri kreiranju objekta definiranje imena, prezimena i godina osobe.

Funkcija koja se nalazi u klasi i ima prvi argument `self`, naziva se metoda. Metoda se poziva nad konkretnim objektom. U ovom primjeru to je `osoba_1`.

U ovom primjeru imamo jednu klasu: `Osoba` i dva objekta: `osoba_1` i `osoba_2`.

#### Dohvaćanje podataka s interneta

U podatkovnoj znanosti važan dio je prikupljanje podataka (prvi korak u *pipeline*-u). Nekad sami moramo prikupiti podatke s interneta. Ovdje će biti prikazana kratka demonstracija pomoću modula `requests`.

```python
import requests as r # Ako modul nije dostupan potrebno je instalirati: `pip install requests`
import xml.etree.ElementTree as ET # Modul za parsiranje XML-a - novi format datoteke.

request = r.get('https://vrijeme.hr/hrvatska_n.xml') # Dohvatimo podatke o vremenu
sadrzaj = request.text
root = ET.fromstring(sadrzaj) # Pretvorimo xml format podataka u objekt

podaci = dict() # Spremit cemo podatke u dictionary

# XML ima hijerarhiju - imamo roditelje i djecu
for child in root:
    if child.tag == "Grad":
        podaci[child.find('GradIme').text] = dict()
        for subchild in child:
            if subchild.tag == "Podatci":
                for podatak in subchild:
                    podaci[child.find('GradIme').text][podatak.tag] = podatak.text

podaci

# Mozemo li umjesto koristenja `dictionary`-a, koristiti klase?
# Kako dizajnirati klasu?
```

#### Baza podataka

Slično kao i kod dohvaćanja podataka s interneta, ponekad podatke moramo pronaći u bazama podataka.
U nastavku demonstrirat ćemo dohvaćanje podataka iz baze podataka pomoću SQL-a (Structured Query Language).

Postoje i NoSQL baze podataka, koje su slične `dictionary` u Pythonu (key-value struktura).

Proučimo SQL sintaksu za `SELECT` naredbu:

`SELECT column1, column2, ... FROM table_name WHERE condition;`

Naredba `SELECT` koristi se kako bi dohvatili podatke iz baze podataka.
Prvo biramo koje stupce želimo dohvatitit (column1, column2, ...), nakon toga navodimo iz koje tablice želimo podatke (table_name) i na kraju uvijete (contition).

```python
import sqlite3 # Ukljucujemo modul za sqlite3 bazu

# Kreiramo koncekciju prema bazi. sqlite koristi datoteku kao spremnik.
conn = sqlite3.connect('weather_data.db')

# Kreiramo cursor objekt, pomocu kojeg komuniciramo s bazom
cursor = conn.cursor()

# Izvrsimo kreiranje tablice.
cursor.execute('''
CREATE TABLE IF NOT EXISTS weather (
    city TEXT,
    date TEXT,
    temperature REAL
)
''')

# Spremimo promjene koje smo napravili u bazu
conn.commit()

# Kreirajmo listu koja sadrzi podatke o gradovima i temperaturi.
data = [
    ('New York', '2024-06-01', 24.0),
    ('Los Angeles', '2024-06-01', 29.0),
    ('Chicago', '2024-06-01', 21.0),
    ('Houston', '2024-06-01', 32.0),
    ('Phoenix', '2024-06-01', 40.0),
    ('Philadelphia', '2024-06-01', 25.5),
    ('San Antonio', '2024-06-01', 35.0),
    ('San Diego', '2024-06-01', 26.5),
    ('Dallas', '2024-06-01', 33.0),
    ('San Jose', '2024-06-01', 25.0),
    ('Austin', '2024-06-01', 34.0),
    ('Jacksonville', '2024-06-01', 31.0),
    ('Fort Worth', '2024-06-01', 32.5),
    ('Columbus', '2024-06-01', 22.0),
    ('San Francisco', '2024-06-01', 20.0),
    ('Charlotte', '2024-06-01', 27.5),
    ('Indianapolis', '2024-06-01', 23.0),
    ('Seattle', '2024-06-01', 18.5),
    ('Denver', '2024-06-01', 29.5),
    ('Washington', '2024-06-01', 26.5)
]

# Ubacimo kreirane podatke u bazu
cursor.executemany('''
INSERT INTO weather (city, date, temperature)
VALUES (?, ?, ?)
''', data)

# Spremimo promjene koje smo napravili u bazu
conn.commit()


# Koristeci DB Browser (SQLite) aplikaciju pogledajmo kako izgleda baza.
```

```python
# Dohvatimo podatke pomocu `SELECT` naredbe.
cursor.execute('SELECT * FROM weather')
rows = cursor.fetchall()

# Kreirajmo listu s rezultatima
data = [row for row in rows]

# Ispišimo listu
print(data)

# Dohvatimo podatke pomocu `SELECT` naredbe, ali sad gradove gdje je temperatura bila veća od 29 stupnjeva.
cursor.execute('SELECT * FROM weather WHERE temperature > 29')
rows = cursor.fetchall()

# Kreirajmo listu s rezultatima
data = [row for row in rows]

# Ispišimo listu
print(data)

# Zatvorimo konekciju prema bazi, slicno kao i s datotekama, zatvaramo resurs
conn.close()
```
