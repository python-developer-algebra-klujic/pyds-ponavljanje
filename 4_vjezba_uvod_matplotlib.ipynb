```python
casedistribution_df = pd.read_csv('data/casedistribution.csv', parse_dates=['dateRep'], dayfirst=True)
casedistribution_df
```

```python
cases_pre_day = casedistribution_df[['dateRep', 'cases']].groupby('dateRep', as_index=False).sum()
cases_pre_day.sort_values(by='dateRep', inplace=True)
cases_pre_day
```

```python
plt.plot(cases_pre_day.dateRep, cases_pre_day.cases)
# plt.show() ovo nije potrebno jer kao i print funkciju Jupyter poziva implicitno
```

```python
plt.plot(cases_pre_day.dateRep, cases_pre_day.cases, 'x-.c')
```
