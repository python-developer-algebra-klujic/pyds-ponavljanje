## Grafovi unutar Pandas biblioteke

```python
mall_customers_df = pd.read_csv('data/mall_customers.csv')
mall_customers_df
```

```python
mall_customers_df.plot()
```

```python
customers_score = mall_customers_df.iloc[:, [2, 4]]
customers_score
```

```python
customers_score_mean = customers_score.groupby(by='Age', as_index=False).mean()
customers_score_mean
```

```python
customers_score_mean.plot(
    kind='line',
    x='Age',
    y='Spending Score (1-100)',
    figsize=(10, 5),
    style='o--b',
    legend=True,
    title='Srednja vrijednost bodovanja po godinama'
)
```
