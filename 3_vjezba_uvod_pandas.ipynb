# Koristenje Pandasa na primjeru Airbnb

```python
import pandas as pd

airbnb_df = pd.read_csv('data/airbnb_new_york_city_listings.csv')
airbnb_df
```

```python
airbnb_df.head(3)
```

```python
airbnb_df.info()
```

```python
airbnb_df['last_review'] = pd.to_datetime(airbnb_df['last_review'])

airbnb_df.dtypes
```

```python
display(airbnb_df.head(3))
display(airbnb_df.tail(3))
```

```python
airbnb_df['last_review_year'] = airbnb_df['last_review'].dt.year

airbnb_df
```

```python
airbnb_df.isnull().sum()
```

- name - Unknown Apartment ili No Name Listing
- host_name - Unknown Host ili Jane Doe
- last_review - 1900-01-01 ili 1970-01-01 "Pocetak UNIX epohe"
- last_review_year - 0
- reviews_per_month - 0.00

```python
pd.DataFrame.fillna?
```

```python
airbnb_df['name'].fillna('Unknown Apartment', inplace=True)
```

```python
airbnb_df['name'] = airbnb_df['name'].fillna('Unknown Apartment')
```

```python
airbnb_df.isnull().sum()
```

```python
airbnb_df['host_name'] = airbnb_df['host_name'].fillna('Jane Doe')
airbnb_df['last_review'] = airbnb_df['last_review'].fillna('1900-01-01')
airbnb_df['last_review_year'] = airbnb_df['last_review_year'].fillna(0)
airbnb_df['reviews_per_month'] = airbnb_df['reviews_per_month'].fillna(0.00)
```

```python
airbnb_df.isnull().sum()
```

```python
airbnb_df.info()
```

```python
airbnb_df['last_review_year'].astype('int64')
```

```python
airbnb_df.info()
```

```python
airbnb_df['last_review_year'] = airbnb_df['last_review_year'].astype('int64')
```

```python
airbnb_df.info()
```

```python
airbnb_df
```

```python
airbnb_df['name'] = airbnb_df['name'].str.strip()
airbnb_df['host_name'] = airbnb_df['host_name'].str.strip()
airbnb_df['neighbourhood_group'] = airbnb_df['neighbourhood_group'].str.strip()
airbnb_df['neighbourhood'] = airbnb_df['neighbourhood'].str.strip()
airbnb_df['room_type'] = airbnb_df['room_type'].str.strip()

airbnb_df
```

```python
airbnb_df.columns
```

```python
airbnb_df['minimum_revenue'] = airbnb_df['price'] * airbnb_df['minimum_nights']

airbnb_df
```

```python
airbnb_df[['room_type', 'price', 'minimum_revenue']].groupby('room_type').mean()
```

```python
airbnb_df[['room_type', 'price', 'minimum_revenue']].groupby('room_type').median()
```

# Korelacija

- 1 - podaci jako koreliraju - ako je puno vremena utroseno na ucenje veca je ocjena
- 0 - podaci nemaju nikakvu korelaciju
- -1 - podaci jako obrnuto koreliraju - ako je puno vremena utroseno na igranje igrica to je manja ocjena

```python
airbnb_df.corr()
```

```python
airbnb_df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'last_review_year']].corr()
```
