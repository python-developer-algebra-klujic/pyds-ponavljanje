## Tipovi podataka u Pandasu

DataFrame i Series

Prije uporabe, potrebno je importirati biblioteku

```python
import pandas as pd
```

```python
students = {
    'first_names': ['Pero', 'Marko', 'Iva', 'Ana'],
    'last_names': ['Peric', 'Markic', 'Ivic', 'Anic'],
    'grades': [15, 45, 89, 78]
}

data_frame = pd.DataFrame(students)
data_frame
```

```python
type(data_frame)
```

```python
stupac = data_frame['first_names']
stupac
```

```python
grades = data_frame['grades']
grades
```

```python
type(grades)
```

```python
names = pd.Series(['Pero', 'Marko', 'Iva', 'Ana'])
names
```

```python
names = pd.Series(['Pero', 'Marko', 'Iva', 'Ana'], index=['Prvi', 'Drugi', 'Treci', 'Cetvrti'])
names
```

```python
names['Drugi']
```

```python
names_dict = {'Prvi': 'Pero', 'Drugi':'Marko', 'Treci':'Iva', 'Cetvrti':'Ana'}
names = pd.Series(names_dict)
names
```

```python
names_part = pd.Series(names, index=['Treci', 'Cetvrti'])
names_part
```
